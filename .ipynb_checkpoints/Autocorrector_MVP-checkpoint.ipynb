## Загрузка библиотек и разбиение текста на токены 

```python
import pandas as pd
import pymorphy3
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter

with open("АннаКоренина.txt", "r", encoding="utf-8-sig") as file:
    text = file.read()
print(text)
tokens = word_tokenize(text.lower())
print(tokens)
print(f"# tokens: {len(word_tokenize(text))}")
```

## Удаление знаков препинания

```python
punctuation_marks = ['!', ',', '(', ')', ':', '-', '?', '.', '..', '...']

only_words = []
for token in tokens:
    if token not in punctuation_marks:
        only_words.append(token)

print(len(only_words))
```

## Удаление стоп-слов

```python
stop_words = stopwords.words("russian")
stop_words
```

```python
filtered_words = []
for token in only_words:
    if token not in stop_words:
        filtered_words.append(token)
filtered_words
```

```python
COUNTS = Counter(filtered_words)
print(COUNTS.most_common(10))
```

```python
def prob_cal(word_count_dict):
    probs = {}
    m = sum(word_count_dict.values())
    for key in word_count_dict.keys():
        probs[key] = word_count_dict[key] / m
    return probs

final_dict = prob_cal(COUNTS)
print(final_dict)
```

```python
from nltk.lm import Vocabulary
from nltk.util import bigrams
from nltk.lm import KneserNeyInterpolated
from nltk.lm.preprocessing import padded_everygram_pipeline

bigram_text = list(bigrams(filtered_words))

vocab = Vocabulary(filtered_words, unk_cutoff=1)

# Генерация паддингов для обработки краевых случаев
train_data, padded_sents = padded_everygram_pipeline(2, bigram_text)

# Инициализация и обучение модели KneserNeyInterpolated с биграммами
lm = KneserNeyInterpolated(2)
lm.fit(train_data, padded_sents, vocab)
```

## Реализация алгоритма Левенштейна

```python
def levenshtein_distance(s, t):
    m = len(s)
    n = len(t)
    d = [[0] * (n + 1) for i in range(m + 1)]  

    for i in range(1, m + 1):
        d[i][0] = i

    for j in range(1, n + 1):
        d[0][j] = j
    
    for j in range(1, n + 1):
        for i in range(1, m + 1):
            if s[i - 1] == t[j - 1]:
                cost = 0
            else:
                cost = 1
            d[i][j] = min(d[i - 1][j] + 1,      # deletion
                          d[i][j - 1] + 1,      # insertion
                          d[i - 1][j - 1] + cost) # substitution   

    return d[m][n]
```

```python
word1 = "слово"
word2 = "слива"
print(levenshtein_distance(word1, word2))
```
